# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4])
plt.xlabel('X Label')
plt.ylabel('funny numbers')
plt.title('Title')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r-.', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {
    'a': np.arange(500),
    'b': np.random.randint(0, 50, 50),
    'c': np.random.randn(500)
    }
plt.scatter('a', 'c', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
